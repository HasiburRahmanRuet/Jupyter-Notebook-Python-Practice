```python
def walk_plot(num):
    choose_set=[1,-1]
    walk=[]
    x=[]
    cur_step=0
    for step in range(1,num+1):
        if step%2==1:
            cur_step=cur_step+choice(choose_set)
        walk.append(cur_step)
        x.append(step)
        
    return x,walk
```

```python
for i in range(1):
    x,y=walk_plot(80)
    plt.plot(x,y, label='Plot No.' + str(i+1))
    plt.title("Random Walk")
    plt.xlabel("Road Map")
    plt.ylabel("Steps")
    plt.legend(loc='lower right')
    
plt.show()

```

```python
def trans_1(x,y):
    return 0.5*x,0.5*y
def trans_2(x,y):
    return 0.5*x+0.5,0.5*y+5
def trans_3(x,y):
    return 0.5*x+1, 0.5*y
```

```python
trans=choice([trans_1,trans_2,trans_3])
```

```python
x=[]
y=[]
a=b=0
for i in range(5000):
    trans=choice([trans_1,trans_2,trans_3])
    a,b=trans(a,b)
    x.append(a)
    y.append(b)
plt.plot(x,y,'o')
plt.show()
    
```
