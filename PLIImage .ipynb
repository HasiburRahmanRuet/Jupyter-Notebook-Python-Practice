```python
f_name="image.jpg"
img=Image.open(f_name)
imshow(img)
```

```python
print(img.format_description)
print(img.size)
print(img.mode)
```

```python
crp_img=img.crop((200,200,600,600))
res_img=img.resize((600,400))
imshow(res_img)
```
